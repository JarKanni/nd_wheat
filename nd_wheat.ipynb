```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Historical Wheat Production

```python
wheat = pd.read_csv("C:/users/jarka/wondering/nd_wheat/nd_hist.csv", index_col=False, thousands=',')

print(wheat.head())

print(wheat.columns)
```

```python

#wheat.plot(kind='line', x='YEAR', y='PLANTED')
#wheat.plot(kind='line', x='YEAR', y='YIELD/ACRE')

year = wheat['YEAR']
planted = wheat['PLANTED']*1000
harvested = wheat['HARVESTED']*1000

plt.plot(year, planted, color='green')
plt.plot(year, harvested, color='gold')

plt.xlabel('Year')
plt.ylabel('Planted Acres (million)')
plt.title('North Dakota Wheat Production 1990-2024')

plt.legend(['Planted','Harvested'])

plt.savefig(fname='ND_wheat_graph.png', format='png')
plt.show()

```
